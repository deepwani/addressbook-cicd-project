```python
# Function to find factorial of given number
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

num = 5
print("Factorial of", num, "is", factorial(num))
```

```python
'find vowels in the string'
def find_vowels(input_str):
  vowels = 0
  for i in input_str:
    if(i in ('a','e','i','o','u')):
      vowels = vowels + 1
  print(vowels)
```

```python
find_vowels( input("Enter a string:"))
```

```python
class Student:
  def __init__(self, name):
    self.name = name
    self.courses = []

  def enroll(self,course):
    print(f"{self.name} is enrolled to {course} successfully")
    self.courses.append(course)

  def withdraw(self,course):
    print(f"{self.name} is withdrawn from {course} successfully")
    self.courses.remove(course)

student1 = Student("XYZ")
student1.enroll('english')
student1.enroll('maths')
student1.enroll('science')
student1.withdraw('maths')
print(student1.courses)
```

```python
class Circle:
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    print(3.14*self.radius*self.radius)

circle1 = Circle(5)
print(circle1.area())
```
